### chat

```python
SYSTEM_PROMPT='You are a helpful assistant. 你是一个乐于助人的助手。'

MODEL_PATH="/content/chinese-alpaca-2-7b/chinese-alpaca-2-7B-Q4_0.gguf"
FIRST_INSTRUCTION="如何能够快速入睡？"

!./llama.cpp/build/bin/llama-cli -m "$MODEL_PATH" \
--color -i -c 4096 -t 8 --temp 0.5 --top_k 40 --top_p 0.9 --repeat_penalty 1.1 \
--in-prefix-bos --in-prefix ' [INST] ' --in-suffix ' [/INST]' -p \
"[INST] <<SYS>>\n$SYSTEM_PROMPT\n<</SYS>>\n\n$FIRST_INSTRUCTION [/INST]"
```
